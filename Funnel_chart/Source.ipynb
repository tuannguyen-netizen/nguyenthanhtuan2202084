```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create a funnel chart from input data
def funnel_chart(labels, values):
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(10, 8))

    # Process x and y positions for each stage of the funnel
    y_pos = np.arange(len(labels))

    # Draw the funnel chart bars
    ax.barh(y_pos, values, color='royalblue', edgecolor='black', height=0.8)

    # Set labels and other properties for the chart
    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels, fontsize=12, fontweight='bold')
    ax.set_xlabel('Quantity', fontsize=14)
    ax.set_title('Funnel Chart Representing the Sales Process', fontsize=16, fontweight='bold')

    # Add values to the bars
    for i, v in enumerate(values):
        ax.text(v + 50, i, str(v), color='black', va='center', fontsize=12)

    # Add gradient coloring effect
    colors = plt.cm.Blues(np.linspace(0.4, 1, len(values)))
    ax.barh(y_pos, values, color=colors, edgecolor='black', height=0.8)

    # Show the chart
    plt.show()

# Extended sample data
labels = [
    'Customer Reach', 'Customer Interest', 'Sign-up for Information',
    'Event Participation', 'Product Demo', 'Price Negotiation',
    'Order Confirmation', 'Payment', 'Customer Retention', 'Referrals'
]

values = [
    2000, 1500, 1200, 900, 700, 500,
    300, 250, 150, 100
]

# Call the function to draw the funnel chart
funnel_chart(labels, values)
```
