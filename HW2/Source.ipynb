```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/drive/MyDrive/CS441_2025/HW2/data/train.csv' # Corrected file path
data = pd.read_csv(file_path)

# Set the visual style
sns.set(style="whitegrid")

# Create subplots for visualizations
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Distribution of SalePrice
sns.histplot(data['SalePrice'], bins=30, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of SalePrice')
axes[0, 0].set_xlabel('SalePrice')
axes[0, 0].set_ylabel('Frequency')

# 2. SalePrice vs LotArea
sns.scatterplot(x=data['LotArea'], y=data['SalePrice'], ax=axes[0, 1], color='green')
axes[0, 1].set_title('SalePrice vs LotArea')
axes[0, 1].set_xlabel('LotArea')
axes[0, 1].set_ylabel('SalePrice')

# 3. SalePrice vs YearBuilt
sns.boxplot(x=data['YearBuilt'], y=data['SalePrice'], ax=axes[1, 0], color='lightcoral')
axes[1, 0].set_title('SalePrice vs YearBuilt')
axes[1, 0].set_xlabel('YearBuilt')
axes[1, 0].set_ylabel('SalePrice')

# 4. SalePrice vs OverallQual
sns.boxplot(x=data['OverallQual'], y=data['SalePrice'], ax=axes[1, 1], color='lightgreen')
axes[1, 1].set_title('SalePrice vs OverallQual')
axes[1, 1].set_xlabel('OverallQual')
axes[1, 1].set_ylabel('SalePrice')

# Adjust layout for better appearance
plt.tight_layout()
plt.show()

# Additional Analysis:
# 5. Correlation heatmap for selected features
correlation = data[['SalePrice', 'LotArea', 'YearBuilt', 'OverallQual']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Features vs SalePrice')
plt.show()

# 6. SalePrice vs Neighborhood (Boxplot to explore neighborhood effect)
plt.figure(figsize=(12, 6))
sns.boxplot(x=data['Neighborhood'], y=data['SalePrice'], palette='Set2')
plt.xticks(rotation=90)
plt.title('SalePrice vs Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('SalePrice')
plt.show()

# 7. SalePrice vs MSZoning (Boxplot to explore zoning effect)
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['MSZoning'], y=data['SalePrice'], palette='Pastel1')
plt.title('SalePrice vs MSZoning')
plt.xlabel('MSZoning')
plt.ylabel('SalePrice')
plt.show()

# 8. SalePrice vs Street (Boxplot to explore street effect)
plt.figure(figsize=(6, 6))
sns.boxplot(x=data['Street'], y=data['SalePrice'], palette='viridis')
plt.title('SalePrice vs Street Type')
plt.xlabel('Street')
plt.ylabel('SalePrice')
plt.show()

# 9. SalePrice vs PoolQC (Boxplot to explore pool quality effect)
plt.figure(figsize=(6, 6))
sns.boxplot(x=data['PoolQC'], y=data['PoolQC'], palette='coolwarm')
plt.title('SalePrice vs PoolQC')
plt.xlabel('Pool Quality')
plt.ylabel('SalePrice')
plt.show()
```
